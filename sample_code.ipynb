```python
# 5001:BEER
# 5002:LIQUOR
# 5003:WINE
def alcohol(data):
    beer = data.query('product_group_code == 5001')
    liquor = data.query('product_group_code == 5002')
    wine = data.query('product_group_code == 5003')
    
    b = beer['purchase_year'].count()
    l = liquor['purchase_year'].count()
    w = wine['purchase_year'].count()
    
    return b,l,w
```

```python
b_03, l_03, w_03 = alcohol(sub_sample03)
b_04, l_04, w_04 = alcohol(sub_sample04)
b_05, l_05, w_05 = alcohol(sub_sample05)
b_06, l_06, w_06 = alcohol(sub_sample06)
b_07, l_07, w_07 = alcohol(sub_sample07)

print("year 2003, beer:", b_03, " liquor:", l_03, " wine:", w_03)
print("year 2004, beer:", b_04, " liquor:", l_04, " wine:", w_04)
print("year 2005, beer:", b_05, " liquor:", l_05, " wine:", w_05)
print("year 2006, beer:", b_06, " liquor:", l_06, " wine:", w_06)
print("year 2007, beer:", b_07, " liquor:", l_07, " wine:", w_07)
print(" ")
print("year 2003, beer/(wine+liquor)=", b_03/(l_03 + w_03))
print("year 2004, beer/(wine+liquor)=", b_04/(l_04 + w_04))
print("year 2005, beer/(wine+liquor)=", b_05/(l_05 + w_05))
print("year 2006, beer/(wine+liquor)=", b_06/(l_06 + w_06))
print("year 2007, beer/(wine+liquor)=", b_07/(l_07 + w_07))
```

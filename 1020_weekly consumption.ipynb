```python
def data_generator(data):
    '''
    # yyyymm: year_month
    # week
    # time_key: year_week
    '''
    data['date'] = pd.to_datetime(data['purchase_date'])
    data['week'] = data['date'].dt.week
    data['time_key'] = data['purchase_year']*100 + data['week']


    total_number = data['upc'].count()
    ounce =[]
    for i in range(total_number):
        if data['size1_units'][i] == 'ML':
            ounce.append(0.033814)
        elif data['size1_units'][i] == 'Li':
            ounce.append(33.814022)
        else:
            ounce.append(1)
    data['unit_ounce'] = ounce


    #q_product: 몇 캔을 샀는가
    #q_ounce: 몇 oz를 마셨는가
    #q_volume: 몇 oz의 에탄올을 섭취했는가 
    data['q_product'] = data['quantity'] * data['multi']
    data['q_ounce'] = data['quantity'] * data['multi'] * data['size1_amount'] * data['unit_ounce']

    if data['product_group_code'][0] == 5001:
        data['q_volume'] = data['quantity'] * data['multi'] * data['size1_amount'] * data['unit_ounce'] * 0.045
    elif data['product_group_code'][0] == 5002:
        data['q_volume'] = data['quantity'] * data['multi'] * data['size1_amount'] * data['unit_ounce'] * 0.411
    else:
        data['q_volume'] = data['quantity'] * data['multi'] * data['size1_amount'] * data['unit_ounce'] * 0.129

    return data
```

```python
def alcohol_sort(data):
    d = {}
    d['alcohol_list'] = data['product_group_code'].unique()
    return pd.Series(d, index=['alcohol_list'])
```

```python
def alcohol_dummy(data):
    total_number = data['alcohol_list'].count()
    beer = []
    liquor = []
    wine = []
    for i in range(total_number):
        if 5001 in data['alcohol_list'][i]:
            beer.append(1)
        else:
            beer.append(0)

        if 5002 in data['alcohol_list'][i]:
            liquor.append(1)
        else:
            liquor.append(0)

        if 5003 in data['alcohol_list'][i]:
            wine.append(1)
        else:
            wine.append(0)


    data['beer'] = beer
    data['wine'] = wine
    data['liquor'] = liquor
    
    return data
```
